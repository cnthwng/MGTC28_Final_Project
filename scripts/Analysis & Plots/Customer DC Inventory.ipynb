# Initial Analysis of Customer DC Inventory CSV
Exploration of data containted in the file utilizing Pandas library and creating plots with Seaborn to identify any potential trends within data 

```python
import pandas as pd
```

```python
df = pd.read_csv('kenvue_data\\Customer DC Inventory - UTSC Lecture.csv', header=[0,1])
```

##### Cleaning & Reorganizing Data for ease of plotting

```python
df = df.set_index([('Fiscal Year  /  Fiscal Week', 'Need State'), ('Unnamed: 1_level_0', 'Unnamed: 1_level_1')])
```

```python
df = df.T
```

```python
df
```

```python
df = df.stack(0)
```

```python
df
```

```python
df = df.reset_index()
```

```python
df
```

```python
df = df.rename_axis(None, axis=1)
```

```python
df = df.rename(columns={df.columns[0]: "year", df.columns[1]: "week", df.columns[2]:"Need States"})
```

```python
df
```

```python
df['date'] = pd.to_datetime(df["year"].astype(str) + "-" + df["week"].astype(str) + "-1", format='%Y-%U-%w')
```

```python
df
```

##### Graphing using the re-organized data

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df
```

```python
State1 = df[(df["Need States"]=="Need State 1")]
State2 = df[(df["Need States"]=="Need State 2")]
State3 = df[(df["Need States"]=="Need State 3")]
State4 = df[(df["Need States"]=="Need State 4")]
State5 = df[(df["Need States"]=="Need State 5")]
```

```python
import warnings
warnings.filterwarnings('ignore')
```

## Plotting Weekly DC Amount 

### Weekly DC Inventory Trends Observed for Each Need State

```python
sns.lineplot(data = State1, x='week', y='Dc Amount', hue='year')
plt.title('Need State 1')
sns.set(rc={'figure.figsize':(14,10)})
plt.xlabel('Week')
plt.ylabel('DC Amount')
plt.legend()
plt.show()
```

```python
sns.lineplot(data = State4, x='week', y='Dc Amount', hue='year')
plt.title('Need State 4')
sns.set(rc={'figure.figsize':(14,10)})
plt.xlabel('Week')
plt.ylabel('DC Amount')
plt.legend()
plt.show()
```
