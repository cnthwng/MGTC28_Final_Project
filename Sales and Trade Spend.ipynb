## Creating a df for Sales & Trade Spend for Each Quarter

```python
# Group by Quarter for Sales

Q1 = df_TSales[(df_TSales['Fiscal Week']<=13)]
Q2 = df_TSales[(df_TSales['Fiscal Week']>13) & (df_TSales['Fiscal Week']<=26)]
Q3 = df_TSales[(df_TSales['Fiscal Week']>26) & (df_TSales['Fiscal Week']<=39)]
Q4 = df_TSales[(df_TSales['Fiscal Week']>39)]
```

```python
Q1 = Q1.groupby('Fiscal Year').sum().reset_index()
Q2 = Q2.groupby('Fiscal Year').sum().reset_index()
Q3 = Q3.groupby('Fiscal Year').sum().reset_index()
Q4 = Q4.groupby('Fiscal Year').sum().reset_index()
```

```python
Q1 = df_TTSpend[(df_TSales['Fiscal Week']<=13)]
Q2 = df_TTSpend[(df_TSales['Fiscal Week']>13) & (df_TSales['Fiscal Week']<=26)]
Q3 = df_TTSpend[(df_TSales['Fiscal Week']>26) & (df_TSales['Fiscal Week']<=39)]
Q4 = df_TTSpend[(df_TSales['Fiscal Week']>39)]
```

```python
Q1 = Q1.groupby('Fiscal Year').sum().reset_index()
Q2 = Q2.groupby('Fiscal Year').sum().reset_index()
Q3 = Q3.groupby('Fiscal Year').sum().reset_index()
Q4 = Q4.groupby('Fiscal Year').sum().reset_index()
```

```python
new = [Q1, Q2, Q3, Q4]
```

```python
new_sheet = pd.concat(new)
new_sheet = new_sheet.sort_values(by=['Fiscal Year'])
new_sheet
```
